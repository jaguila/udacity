Here is a quick example of an entity details API call:

http://api.followthemoney.org/entity.php?eid=#######&APIKey=XXXXXXXX&mode=xml
replace ##### with your desired Entity ID

replace XXXXXX with your API key

mode=xml can be replaced by mode=json if you desire json output.

```python
df=pd.read_csv('C:/pyproj/data.csv')

df['total_byk']=df.Total.apply(lambda x: x/1000)
df.head()
```

```python
grp=['elected_official', 'Industry']
```

```python
import matplotlib.pyplot as plt
# %matplotlib inline
from matplotlib.figure import Figure
f = Figure( figsize =(7,7) )


# plt.pyplot.close(f)
```

```python
def select(lastname):
    df1=df[df['last']==lastname]
    df_final=df1[['elected_official', 'Industry', 'total_byk']].groupby(grp)
    df_final=df_final.sum().reset_index()
    return df_final
```

```python
def pie():
    df_grph=select('BROWN JR')


    n=len(df_grph['Industry'])
    lab=[' ']*n
    df_grph.plot(kind="pie", x='Industry', y='total_byk', labels=lab, autopct='%1.0f%%', pctdistance=1.1, labeldistance=1.2)
    plt.legend(df_grph['Industry'], loc='center left', bbox_to_anchor=(1.0, 0.5))
    plt.xlabel("percentage of campaign funds")
    plt.ylabel('')

    plt.show()
```

```python
pie()
```
